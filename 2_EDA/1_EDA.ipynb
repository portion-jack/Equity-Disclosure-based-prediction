## EDA

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'AppleGothic'
sns.set(font="AppleGothic", 
        rc={"axes.unicode_minus":False},
        style='darkgrid')

import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv("../data/preprocessed_data/data_(main).csv", encoding='cp949', index_col=0)
```

```python
eda_data = pd.read_csv("../data/preprocessed_data/forEDA.csv", encoding='cp949', index_col=0)
```

```python
palette1 = sns.color_palette('colorblind', 10)
sns.palplot(palette1)
```

```python
eda_data["보고구분"].value_counts().plot.pie(autopct='%.1f%%')
plt.title("지분공시 보고구분 비율")
plt.ylabel("보고구분")
plt.show()
```

```python
cols = list(eda_data.dtypes[eda_data.dtypes == "object"].index)[1:]
cols
```

```python
for col in cols:
    print("="*50, "\n")
    print(eda_data[col].value_counts())
    print("\n카테고리 갯수 :",len(eda_data[col].value_counts()), "\n")
```

```python
sns.countplot(x='보고구분', data=eda_data, palette=palette1)
```

```python
sns.countplot(x='직위명', data=eda_data, palette=palette1)
```

```python
sns.countplot(x='보고자 구분', data=eda_data, palette=palette1)
```

```python
sns.countplot(x='요일', data=eda_data,
              palette=palette1, order=eda_data["요일"].value_counts().index).set(title='요일별 지분공시 수')
```

```python
eda_data["임원"].value_counts().plot.pie(autopct='%.1f%%')
plt.title("임원 구성비율")
plt.ylabel("임원")
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.countplot(x='변동사유', data=eda_data,
                   palette=palette1,
                   order=eda_data["변동사유"].value_counts().index).set(title='변동사유 별 지분공시 수')
plt.xticks(rotation=45)
```
